Снятие выборки

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from preprocessing import get_image
from grains import collect_grains, get_best_distribution, show_best_fit, best_fit_moments
import warnings
warnings.filterwarnings("ignore")

image = get_image("data/structures/S2.png")
# fig, ax = plt.subplots()
# plt.imshow(image)

grains, cuts_illustration = collect_grains(image)
fig, ax = plt.subplots()
plt.imshow(cuts_illustration, cmap='gray')
fig, ax = plt.subplots()
grains = np.array(grains)
bins = max(grains) + 1
plt.hist(grains, density=False, bins=bins)
```
